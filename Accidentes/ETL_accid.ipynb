```python
# Se realiza la importación de librerías:
import pandas as pd
```

```python
# Se abre el archivo:
with open("Bases_datos/homicidios.xlsx", 'rb') as f:
    data = pd.read_excel(f)
```

```python
# Muestra de la data:
data.head(5)
```

```python
# Verificar los nulos de cada columna, por la info y tipo de dato:
data.info()
```

```python
# Eliminar columnas en dataset
col_to_drop = ["XY (CABA)", "LUGAR_DEL_HECHO", "TIPO_DE_CALLE", "Calle", "Altura", "Cruce"]
data.drop(columns=col_to_drop, inplace=True)
```

```python
data.info()
```

```python
# Replace column names
data.columns = data.columns.str.replace("ID", "Id_Fecha")
data.columns = data.columns.str.replace("N_VICTIMAS", "Num_Victs")
data.columns = data.columns.str.replace("FECHA", "Fecha")
data.columns = data.columns.str.replace("AAAA", "Agno")
data.columns = data.columns.str.replace("AAAA", "Agno") 
data.columns = data.columns.str.replace("MM", "Mes")
data.columns = data.columns.str.replace("DD", "Dia") 
data.columns = data.columns.str.replace("HORA", "Hora") 
data.columns = data.columns.str.replace("HH", "Horario") 
data.columns = data.columns.str.replace("Dirección Normalizada", "Direccion")
data.columns = data.columns.str.replace("COMUNA", "Comuna")
data.columns = data.columns.str.replace("PARTICIPANTES", "Participantes")
data.columns = data.columns.str.replace("VICTIMA", "Victima")
data.columns = data.columns.str.replace("ACUSADO", "Acusado")
```

```python
# Guardar el df en formato CSV para proceder al EDA:
data.to_csv('Accid.csv', index=False)
```
